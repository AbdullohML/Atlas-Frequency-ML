```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.pipeline import Pipeline
import submission_helper

df_train = pd.read_csv("datasets/task2_train.csv")
X_train = df_train.drop(columns=["y"]).values
y_train = df_train["y"].values

selector = SelectFromModel(
    RandomForestClassifier(n_estimators=200, random_state=42),
    max_features=50,
    threshold=-np.inf
)

pipeline = Pipeline([
    ('feature_selection', selector),
    ('classifier', RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1))
])

pipeline.fit(X_train, y_train)

selected_mask = selector.get_support()
selected_features = np.where(selected_mask)[0]

importances = pipeline.named_steps['classifier'].feature_importances_
top_indices = np.argsort(importances)[-10:][::-1]

submission_helper.save_task2_model(pipeline, "task2_model.pkl")
```
